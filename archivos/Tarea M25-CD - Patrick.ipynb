```python
import math

# Ejercicio 1A
# Obtener la raíz cuadrada de un número SIN utilizar lambda
def raiz_cuadrada(x):
    return math.sqrt(x)

numero = 25
print("Raíz cuadrada sin lambda:", raiz_cuadrada(numero))


# Ejercicio 1B
# Obtener la raíz cuadrada de un número UTILIZANDO lambda
raiz_lambda = lambda x: math.sqrt(x)

numero = 25
print("Raíz cuadrada con lambda:", raiz_lambda(numero))
```

```python
# Ejercicio 2A
# Obtener el largo de cada palabra SIN utilizar map
texto = "Python es un lenguaje poderoso para datos"
palabras = texto.split()

longitudes = []

for palabra in palabras:
    longitudes.append(len(palabra))

print("Longitudes sin map:", longitudes)


# Ejercicio 2B
# Obtener el largo de cada palabra UTILIZANDO map
texto = "Python es un lenguaje poderoso para datos"
palabras = texto.split()

longitudes = list(map(len, palabras))

print("Longitudes con map:", longitudes)
```

```python
# Ejercicio 3A
# Calcular el producto de una lista SIN utilizar reduce
numeros = [2, 3, 4, 5]

producto = 1

for n in numeros:
    producto *= n

print("Producto sin reduce:", producto)


# Ejercicio 3B
# Calcular el producto de una lista UTILIZANDO reduce
from functools import reduce

numeros = [2, 3, 4, 5]

producto = reduce(lambda x, y: x * y, numeros)

print("Producto con reduce:", producto)
```

```python
# Ejercicio 4A
# Filtrar palabras SIN utilizar filter
palabras = ["python", "Data1", "ciencia", "AI", "modelo3", "codigo"]

resultado = []

for palabra in palabras:
    if any(c.isupper() for c in palabra) or any(c.isdigit() for c in palabra):
        resultado.append(palabra)

print("Resultado sin filter:", resultado)


# Ejercicio 4B
# Filtrar palabras UTILIZANDO filter
palabras = ["python", "Data1", "ciencia", "AI", "modelo3", "codigo"]

resultado = list(filter(lambda palabra: any(c.isupper() for c in palabra) or any(c.isdigit() for c in palabra), palabras))

print("Resultado con filter:", resultado)
```
