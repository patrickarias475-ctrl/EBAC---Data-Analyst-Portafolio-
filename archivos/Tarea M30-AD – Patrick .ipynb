```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Establece el formato de visualizaciÃ³n para nÃºmeros flotantes
pd.options.display.float_format = '{:,.2f}'.format

# Define la ruta del archivo en la carpeta de descargas del usuario
ruta_archivo = os.path.join(r'C:\Users\patri\Downloads', 'fifa_eda.csv')

# Carga el archivo con la codificaciÃ³n y tipos de datos especificados
df = pd.read_csv(ruta_archivo, encoding='latin1', dtype={'ITEM_CODE': str})

# Realiza el diagnÃ³stico inicial obligatorio
print("--- DiagnÃ³stico Inicial: Primeras Filas ---")
print(df.head())
print("\n--- DiagnÃ³stico Inicial: Nombres Exactos de Columnas ---")
print(df.columns.tolist())
```

```python
# Muestra las primeras 20 filas del archivo
display(df.head(20))

# Muestra las Ãºltimas 5 filas
display(df.tail(5))

# Muestra un sample aleatorio de 10 filas
display(df.sample(10))
```

```python
# Muestra los tipos de datos y la estadÃ­stica descriptiva inicial
print(df.info())

# Define una funciÃ³n para limpiar valores monetarios (ej. â¬100M -> 100000000)
def limpiar_finanzas(valor):
    if isinstance(valor, str):
        valor = valor.replace('â¬', '').replace('K', '*1e3').replace('M', '*1e6')
        try:
            return eval(valor)
        except:
            return 0
    return valor

# Pasa a numÃ©ricas las columnas financieras para incluirlas en el describe()
df['Wage'] = df['Wage'].apply(limpiar_finanzas)
df['Release Clause'] = df['Release Clause'].apply(limpiar_finanzas)

# Genera la data estadÃ­stica actualizada
display(df.describe())
```

```python
# Convierte la columna Joined a numÃ©rica (ya que contiene solo el aÃ±o)
df['Joined_Year'] = pd.to_numeric(df['Joined'], errors='coerce')

# AÃ±ade la columna "Years Playing", calcula el aÃ±o actual (2026) menos el ingreso
# Realiza el cÃ¡lculo de antigÃ¼edad para cada jugador
df['Years Playing'] = 2026 - df['Joined_Year']

# Muestra el resultado de la nueva columna
display(df[['Name', 'Joined', 'Years Playing']].head())
```

```python
# Busca y muestra a todos los jugadores de MÃ©xico
jugadores_mexico = df[df['Nationality'] == 'Mexico']
display(jugadores_mexico)

# Ordena y muestra los datos por la columna Release Clause (sueldo/clÃ¡usula) de forma descendente
df_ordenado = df.sort_values(by='Release Clause', ascending=False)
display(df_ordenado.head(10))
```

```python
# Genera un nuevo dataset que contiene el aÃ±o (joined) y el nÃºmero de jugadores
# Agrupa los datos por aÃ±o de ingreso
resumen_anual = df.dropna(subset=['Joined_Year']).groupby('Joined_Year').size().reset_index(name='Num_Players')
display(resumen_anual)

# Genera un grÃ¡fico que contiene, por aÃ±o, el nÃºmero de jugadores
plt.figure(figsize=(12, 6))
plt.bar(resumen_anual['Joined_Year'], resumen_anual['Num_Players'], color='#5dade2', edgecolor='black')

# Configura el eje X para mostrar los aÃ±os correctamente sin notaciÃ³n cientÃ­fica
plt.xticks(resumen_anual['Joined_Year'], rotation=45)

plt.title('DistribuciÃ³n de Jugadores por AÃ±o de Ingreso', fontsize=14)
plt.xlabel('AÃ±o de ContrataciÃ³n')
plt.ylabel('Cantidad de Jugadores')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()

# Guarda y muestra el grÃ¡fico
plt.show()
```
