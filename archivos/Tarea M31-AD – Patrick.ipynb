```python
import pandas as pd
import os
import numpy as np

# Configura el formato general para los valores flotantes
pd.options.display.float_format = '{:,.2f}'.format

# Construye la ruta al archivo en la carpeta de descargas
ruta_archivo = os.path.join(r"C:\Users\patri\Downloads", "fifa_eda.csv")

# Lee el archivo desde la ruta especificada
df = pd.read_csv(ruta_archivo, encoding='latin1', dtype={'ITEM_CODE': str})

# Diagnóstico de verificación inmediata
# 1. Muestra los primeros datos
print(df.head())
# 2. Muestra los nombres EXACTOS de las columnas
print(df.columns.tolist())
```

```python
# Obtiene una muestra aleatoria de 20 líneas para inspección
muestra_20 = df.sample(n=20)
print("\n--- Muestra de 20 líneas ---")
print(muestra_20)

# Selecciona columnas numéricas de interés
columnas_interes = ['Overall', 'Potential'] 

# Genera estadísticas globales (Promedio, Mínimo y Máximo)
stats_globales = df[columnas_interes].agg(['mean', 'min', 'max'])
print("\n--- Estadísticas Globales ---")
print(stats_globales)

# Genera estadísticas agrupadas por nacionalidad
stats_por_pais = df.groupby('Nationality')[columnas_interes].agg(['mean', 'min', 'max'])
print("\n--- Estadísticas por País (primeros 5) ---")
print(stats_por_pais.head())
```

```python
# Crea un DataFrame con jugadores de prueba
jugadores_ficticios = pd.DataFrame([
    {'Name': 'Jugador Ficticio 1', 'Nationality': 'Mexico', 'Overall': 99.0, 'Potential': 99.0},
    {'Name': 'Jugador Ficticio 2', 'Nationality': 'Brazil', 'Overall': 15.0, 'Potential': 20.0}
])

# Une los nuevos registros al DataFrame original
df = pd.concat([df, jugadores_ficticios], ignore_index=True)

# Verifica la inserción visualizando el final del archivo
print("\n--- Últimos registros tras añadir jugadores ficticios ---")
print(df.tail())

# Recalcula estadísticas para observar el cambio tras la inserción
stats_globales_nuevas = df[columnas_interes].agg(['mean', 'min', 'max'])
print("\n--- Nuevas Estadísticas Globales (Comparación) ---")
print(stats_globales_nuevas)
```

```python
# Diccionario de referencia para mapeo
diccionario_geografico = {
    'Mexico': {'Capital': 'Ciudad de México', 'Zona': 'UTC-6'},
    'Brazil': {'Capital': 'Brasilia', 'Zona': 'UTC-3'},
    'Argentina': {'Capital': 'Buenos Aires', 'Zona': 'UTC-3'},
    'Spain': {'Capital': 'Madrid', 'Zona': 'UTC+1'}
}

# Aplica funciones lambda para asignar Capital y Zona Horaria según la nacionalidad
df['Capital'] = df['Nationality'].apply(lambda x: diccionario_geografico.get(x, {}).get('Capital', 'No Registrada'))
df['Zona_Horaria'] = df['Nationality'].apply(lambda x: diccionario_geografico.get(x, {}).get('Zona', 'No Registrada'))

# Aplica función lambda para identificar candidatos al Balón de Oro (Potential > 85)
df['Premio'] = df['Potential'].apply(lambda x: 'Candidato' if pd.notna(x) and x > 85 else 'Next Year')

# Muestra el resultado final de las nuevas columnas
print("\n--- Resultado Final: Columnas Geográficas y Premios ---")
print(df[['Name', 'Nationality', 'Capital', 'Zona_Horaria', 'Premio']].tail(10))
```
