```python
import pandas as pd
import os
import json

# Configuración general
pd.options.display.float_format = '{:,.2f}'.format
ruta_base = r'C:\Users\patri\Downloads'
```

```python
# --- 1. Archivo CSV ---
ruta_csv = os.path.join(ruta_base, 'fifa_eda.csv')

# Carga del archivo
df_csv = pd.read_csv(ruta_csv, encoding='latin1', dtype={'ITEM_CODE': str})

# Diagnóstico de verificación
# 1. Para ver los datos
print("--- Diagnóstico CSV ---")
print(df_csv.head())
# 2. Para ver los nombres EXACTOS de las columnas (sin recortes)
print("\nColumnas CSV:")
print(df_csv.columns.tolist())

# Guardar en una lista (lista anidada)
lista_csv = df_csv.values.tolist()

# Obtener las 10 primeras y 10 últimas filas
primeras_10_csv = df_csv.head(10)
ultimas_10_csv = df_csv.tail(10)

# Seleccionar de 2 a 4 columnas y generar un archivo nuevo
# Se seleccionan las primeras 3 columnas a modo de ejemplo
subset_csv = df_csv.iloc[:, :3] 
subset_csv.to_csv(os.path.join(ruta_base, 'fifa_subset_creado.csv'), index=False, encoding='latin1')
```

```python
import zipfile

# --- 3. Archivo JSON (Desde archivo ZIP) ---
ruta_zip = os.path.join(ruta_base, 'GOT Subtitles.zip')

# Extraer el contenido del ZIP en la misma carpeta de descargas
with zipfile.ZipFile(ruta_zip, 'r') as zip_ref:
    zip_ref.extractall(ruta_base)

# Seleccionamos el primer archivo extraído para el ejercicio
ruta_json = os.path.join(ruta_base, 'season1.json')

# Carga del archivo
df_json = pd.read_json(ruta_json, encoding='latin1', dtype={'ITEM_CODE': str})

# 1. Para ver los datos
print(df_json.head())
# 2. Para ver los nombres EXACTOS de las columnas (sin recortes)
print(df_json.columns.tolist())

# Guardar en una lista (lista de diccionarios para JSON)
lista_json = df_json.to_dict('records')

# Obtener las 10 primeras y 10 últimas filas
primeras_10_json = df_json.head(10)
ultimas_10_json = df_json.tail(10)

# Seleccionar de 2 a 4 columnas y generar un archivo nuevo
# Nota: Ajusta los índices [:, :2] dependiendo de cuántas columnas reales tenga tu JSON tras extraerlo
subset_json = df_json.iloc[:, :2] 
subset_json.to_json(os.path.join(ruta_base, 'GOT_season1_subset.json'), orient='records', force_ascii=False)
```
