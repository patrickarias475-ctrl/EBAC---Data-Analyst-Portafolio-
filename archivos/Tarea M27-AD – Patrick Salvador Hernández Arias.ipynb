```python
import pandas as pd
import os

# 1. Formato General
pd.options.display.float_format = '{:,.2f}'.format

# Ruta del archivo que NO existe
ruta_archivo = os.path.join(r"C:\Users\patri\Downloads", "test.txt")

try:
    print(f"Buscando el archivo en: {ruta_archivo}")
    
    # Intentando aplicar el protocolo de lectura
    df = pd.read_csv(ruta_archivo, encoding='latin1', dtype={'ITEM_CODE': str})
    
    # 1. Para ver los datos
    print(df.head())
    # 2. Para ver los nombres EXACTOS de las columnas (sin recortes)
    print(df.columns.tolist())

except FileNotFoundError:
    # Capturando la excepción específica si el archivo no existe
    print("-> EXCEPCIÓN: El archivo 'test.txt' no se encontró en el directorio especificado.")

except Exception as e:
    # Capturando cualquier otro error inesperado
    print(f"-> ERROR INESPERADO: Ocurrió un problema diferente: {e}")

else:
    # Se ejecuta solo si el bloque 'try' fue exitoso
    print("-> ÉXITO: El archivo se cargó y los diagnósticos se ejecutaron correctamente.")

finally:
    # Se ejecuta SIEMPRE, haya error o no
    print("-> FINALLY: Finalizando el bloque de control de errores de la carga del archivo.")
```
